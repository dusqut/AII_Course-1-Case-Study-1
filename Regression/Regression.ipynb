```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([4, 7, 11, 19, 30])
```

### Linear regression

![LR](https://files.realpython.com/media/fig-lin-reg.a506035b654a.png)

```python
linear_reg = LinearRegression()
linear_reg.fit(X, y)
linear_pred = linear_reg.predict(X)
```

### Polynomial regression

```python
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)  # Perform the linear regression with the polynomial features (X_poly)
poly_pred = poly_reg.predict(X_poly)
```

### Evaluation metrics

#### Mean Squared Error (MSE):

* MSE measures the average squared difference between the predicted values and the actual values in a regression model.
* It is calculated by taking the average of the squared differences between the predicted and actual values.
* MSE provides a measure of how well the model fits the data, with lower values indicating better fit.
* The formula for MSE is: MSE = (1/n) * Î£(y_actual - y_predicted)^2, where n is the number of data points, y_actual is the actual value, and y_predicted is the predicted value.
* MSE is sensitive to outliers because the squared differences magnify larger errors.

#### R-squared:

* R-squared, also known as the coefficient of determination, measures the proportion of the variance in the dependent variable that is predictable from the independent variables.
* It represents the goodness of fit of the model, indicating how well the model explains the variation in the data.
* R-squared ranges from 0 to 1, where 0 indicates that the model does not explain any of the variance in the data, and 1 indicates a perfect fit.
* The formula for R-squared is: R^2 = 1 - (SSR/SST), where SSR is the sum of squared residuals and SST is the total sum of squares.
* R-squared can be interpreted as the percentage of the dependent variable's variation that is explained by the independent variables.

```python
linear_mse = mean_squared_error(y, linear_pred)
linear_r2 = r2_score(y, linear_pred)
poly_mse = mean_squared_error(y, poly_pred)
poly_r2 = r2_score(y, poly_pred)
```

```python
# Printing evaluation metrics
print(f"Linear Regression MSE: {linear_mse:.2f}")
print(f"Linear Regression R-squared: {linear_r2:.2f}")
print(f"Polynomial Regression MSE: {poly_mse:.2f}")
print(f"Polynomial Regression R-squared: {poly_r2:.2f}")
```

```python
# Plotting the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, linear_pred, color='red', label='Linear Regression')
plt.plot(X, poly_pred, color='green', label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear vs Polynomial Regression')
plt.legend()
plt.show()
```

### SVR

Support Vector Regression (SVR) is a variation of Support Vector Machines (SVM) that is used for regression tasks. SVR is a supervised learning algorithm that aims to find a function that best fits the training data while also minimizing the prediction error.

1. Choosing a Kernel:

SVR uses a kernel function to transform the input data into a higher-dimensional space.
Commonly used kernel functions include linear, polynomial, radial basis function (RBF), and sigmoid.
The choice of kernel depends on the nature of the data and the problem at hand.

2. Building the SVR Model:

SVR aims to find a function that best fits the training data while allowing a certain margin of error.
The model seeks to find a hyperplane that maximizes the margin around the predicted values.
The model is trained by solving an optimization problem to find the coefficients that define the hyperplane.

3. Training the Model:

The SVR model is trained by finding support vectors, which are the data points closest to the hyperplane.
The model learns the optimal coefficients by minimizing the error between the predicted and actual values, while also considering a regularization term to control the complexity of the model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
# SVR regression
svr = SVR(kernel='linear')
svr.fit(X, y)
svr_pred = svr.predict(X)
```

```python
# Evaluation
svr_mse = mean_squared_error(y, svr_pred)
svr_r2 = r2_score(y, svr_pred)
```

```python
# Plotting the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, svr_pred, color='red', label='SVR Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```

#### SVR Kernel
The kernel parameter is set to 'linear' to perform linear regression using SVR. You can experiment with other kernel functions like 'poly' or 'rbf' to explore different non-linear regression techniques.

```python
# SVR regression with 'poly' kernel
svr = SVR(kernel='poly')
svr.fit(X, y)
svr_pred = svr.predict(X)
```

```python
# Evaluation
svr_mse = mean_squared_error(y, svr_pred)
svr_r2 = r2_score(y, svr_pred)
print(f"SVR Regression MSE: {svr_mse:.2f}")
print(f"SVR Regression R-squared: {svr_r2:.2f}")
```

```python
# Plotting the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, svr_pred, color='red', label='SVR Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
